```python
# poetry remove langchain-upstage
# poetry add pypdf
import os
import json
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

# PDF 파일 경로 설정
pdf_filepath = 'data/tutorial-korean.pdf'

# 파일 존재 여부 확인 (파일이 없으면 오류 발생)
if not os.path.exists(pdf_filepath): # 파일이 존재하지 않으면
    raise FileNotFoundError(f"파일을 찾을 수 없습니다: {pdf_filepath}") # FileNotFoundError() 발생
```

```python
try:
    # 1. PDF 파일 로드
    # PDF 파일을 로드할 객체 생성
    loader = PyPDFLoader(pdf_filepath) # PyPDFLoader 객체를 반환해서 loader 변수 메모리에 할당
    docs = loader.load()  # PyPDFLoader 객체를 Document 객체(또는 객체 리스트)로 변환

    # 총 문서 개수 출력
    print(f"총 {len(docs)}개의 문서가 로드 되었습니다.")
    print(f"문서 타입: {type(docs[0])}\n")

    #  첫 번째 문서의 메타데이터 출력
    print("첫 번째 문서 메타데이터:")
    print(
        json.dumps(             # JSON 문서로 변환시켜라
            docs[0].metadata,   # Document 객체인 docs[0]의 metadata를
            indent=2,           # 띄어쓰기 2개로 들여쓰기를 적용하고
            ensure_ascii=False  # 한글, 특수문자 등을 아스키 코드로 변환하지 말고 그대로 출력해라
            )
        )

    # 특정 인덱스(10번째) 문서의 내용 확인 (존재할 경우)
    if len(docs) > 10:
        print("\n10번째 문서 내용:", type(docs[10]))
        print(docs[10])  # 10번째 문서 출력




    # 2. 텍스트 분할 (200자 단위, 중첩 없음)
    # RecursiveCharacterTextSplitter()를 사용한 분할기 생성(최대 글자 수: 200개, 중복 단어 수: 0개)
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=200, chunk_overlap=0)
    # split_docs = loader.load_and_split(text_splitter=text_splitter)는 내부적으로 이러한 과정을 거침
        # 1) loader.load()를 통해 전체 문서를 로드 (PyPDFLoader 책임) 
        # all_docs = loader.load()
        # 2) 로드 된 문서를 TextSplitter를 사용하여 분할 (TextSplitter 책임)
        # split_docs = text_splitter.split_documents(all_docs)
    split_docs = loader.load_and_split(text_splitter=text_splitter)  # 분할기를 사용하여 문서 로드

    # 분할된 문서 개수 출력
    print()
    print(f"분할된 문서의 개수: {len(split_docs)}")
    print(f"분할 된 문서들의 타입: {type(split_docs)}")
    print(f"하나의 문서 타입: {type(split_docs[0])}")
    print()
    
    # 10번째 분할된 문서 내용 출력 (존재할 경우)
    if len(split_docs) > 10: # 문서의 길이가 10보다 클 경우
        print("\n10번째 분할된 문서:")
        print(split_docs[10])



    # 3. Lazy Load 방식으로 문서 로드
    print("\nLazy Load 방식으로 문서 로드:")
    for i, doc in enumerate(loader.lazy_load()):
        if i < 5:  # 너무 많은 출력 방지 (예제: 처음 5개만 출력)
            print(json.dumps(doc.metadata, indent=2, ensure_ascii=False))

except Exception as e:
    # 오류 발생 시 메시지 출력
    print(f"오류 발생: {e}")
```

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import FAISS
from langchain_openai import OpenAIEmbeddings, ChatOpenAI
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate

print("==> 1. 문서 로딩 → PDF 읽기...")
loader = PyPDFLoader('./data/tutorial-korean.pdf') # pdf 문서를 읽어와서 PyPDFLoader 클래스 타입의 인스턴스 객체로 변환 후 loader에 저장
documents = loader.load()   # loader 인스턴스 객체를 Document 타입의 list로 변경
print(f"  총 {len(documents)}페이지 로드 완료")


print("==> 2. 문서 분할 → 작은 청크로 나누기")
text_splitter = RecursiveCharacterTextSplitter( # RecursiveCharacterTextSplitter() 생성자를 사용해서 분할기 생성
    chunk_size=1000,        # 청크 크기 (한국어 최적화) (1000자 이하로만 분할하기)
    chunk_overlap=200,      # 중복 부분 (맥락 보존) (중복 단어 200자 허용)
    separators=["\n\n", "\n", ".", " ", ""] # 자연스러운 분할을 위한 구분자
)
chunks = text_splitter.split_documents(documents) # 분할기 적용

print(f"  {len(chunks)}개 청크 생성 완료")
print(f"  평균 청크 길이: {sum(len(chunk.page_content) for chunk in chunks) / len(chunks):.0f}자")

print("==> 3. 벡터화 → 임베딩으로 변환")
embeddings = OpenAIEmbeddings(
    model="text-embedding-3-large", # 고성능 임베딩 모델
    dimensions=1536                 # 임베딩 백터의 차원(크기) 정의하기
)

print("==> 4. 저장 → FAISS 벡터스토어에 저장")
# 'chunks'의 텍스트들을 'embeddings' 모델(text-embedding-3-large, 1536차원)을 사용하여 임베딩 벡터로 변환 후 vectorstore에 저장
vectorstore = FAISS.from_documents(documents=chunks, embedding=embeddings)
print(f" FAISS 벡터스토어 생성 완료 ({len(chunks)}개 벡터)")

print("===> 5. 검색 → 질문과 유사한 문서 찾기")
retriever = vectorstore.as_retriever(
    search_type="similarity",
    search_kwargs={"k": 6}  # 상위 6개 관련 문서 검색
)
print(" Retriever 설정 완료")

print("===> 6. 생성 → LLM으로 답변 생성")
llm = ChatOpenAI(
    model="gpt-4o",
    temperature=0.1,
    max_tokens=1500     # 최대 1500 토큰만 생성해라
)

# 한국어 최적화 프롬프트
prompt_template = """
당신은 BlueJ 프로그래밍 환경 전문가입니다. 
아래 문서 내용을 바탕으로 정확하고 친절한 답변을 제공해주세요.

문서 내용:
{retrieveDocuments}

질문: {question}

답변 규칙:
1. 문서 내용만을 근거로 답변하세요
2. 단계별 설명이 필요하면 순서대로 작성하세요  
3. 구체적인 메뉴명, 버튼명을 포함하세요
4. 문서에 없는 정보는 "문서에서 찾을 수 없습니다"라고 하세요

답변:"""

prompt = PromptTemplate(
    template=prompt_template,
    input_variables=["retrieveDocuments", "question"]
)
print(" 프롬프트 설정 완료")

# ===================================
# 7. QA 체인 생성
# ===================================
print("\n ===> 7.  QA 체인 생성...")
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,            # 모델
    # 검색 된 여러 문서(청크)를 어떠한 방식으로 전달(stuffing)할 지 결정하는 전략 (default="stuff")
    # "stuff"방식: 검색 된 모든 문서(청크)의 내용을 하나의 긴 문자열로 합쳐서 LLM의 프롬프트에 통째로 넣어주는 방식
        # context = "\n\n.join( ... )"와 유사
    chain_type="stuff",     # 리트리버를 그대로 가져와라
    retriever=retriever,    # 사용자 질문에 대해 관련 문서를 검색할 때 사용할 Retriever 객체 지정
    chain_type_kwargs={
        "prompt": prompt,   # llm에게 최종 질문을 전달할 때 사용할 PromptTemplate 지정하기
        # prompt 템플릿 내에서 검색 된 문서 내용이 들어 갈 플레이스 홀더의 이름("retrieveDocuments")
        "document_variable_name": "retrieveDocuments" 
    },
    # return_source_documents 매개변수가 True 이기 때문에
    #   llm이 최종 생성한 답변 외 답변 생성에 사용 된 참조 문서도 함께 반환하도록 함
    return_source_documents=True 
)
print("  RAG 파이프라인 구축 완료!")

# ===================================
# 8. 테스트 질문들
# ===================================
test_questions = [
    "BlueJ에서 객체를 생성하는 방법은 무엇인가요?",
    "컴파일 오류가 발생했을 때 어떻게 확인할 수 있나요?", 
    "디버깅을 위해 중단점을 설정하는 방법을 알려주세요",
    "코드패드는 무엇이고 어떻게 사용하나요?",
    "애플릿을 만들고 실행하는 방법을 설명해주세요"
]

print("\n" + "=" * 60)
print(" RAG 시스템 테스트")
print("=" * 60)

# ===================================
# 9. 질문 및 답변 실행
# ===================================
for i, question in enumerate(test_questions, 1):
    print(f"\n【테스트 {i}/5】")
    print(f" 질문: {question}")
    print(" 답변 생성 중...")
    
    # RAG 실행
    # RetrievalQA.keys(): 
    # dict_keys([
        # 'query',  # 사용자 원본 질문
        # 'result', # llm이 생성한 최종 답변 문자열
        # 'source_documents'    # return_source_documents=True이기에 답변을 생성하는데 사용된 원본 Document 객체들의 리스트 6개
        # ])
    result = qa_chain.invoke({"query": question})
    answer = result["result"]
    source_docs = result["source_documents"]
    
    print(f"\n 답변:")
    print("-" * 50)
    print(answer)
    
    # 참조 문서 정보
    print(f"\n 참조 문서:")
    for j, doc in enumerate(source_docs[:3], 1):
        # 각 Document 객체의 메타데이터에서 'page'값을 가져오고 없으면 'N/A'로 설정
        page = doc.metadata.get('page', 'N/A')
        # 각 Document 객체의 page_content에서 처음 80글자를 가져와 줄바꿈을 공백으로 대체 (미리보기)
        preview = doc.page_content[:80].replace('\n', ' ')
        # 출력
        print(f"   {j}. 페이지 {page}: {preview}...")
    
    print("\n" + "-" * 40)
```
